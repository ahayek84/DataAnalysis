## 1 Perumtations and Combination 
When the order doesn't matter, it is a Combination.
When the order does matter it is a Permutation.
best source :
https://www.mathsisfun.com/combinatorics/combinations-permutations.html

1. Permutations with Repetition
n × n × ... (r times) = nr
2.Permutations without Repetition 
n! = n × (n -1) × (n - 2) ×  .. .. 1
3.if we want to choose a certain number (r) out of all permutations n
nPr = n!/(n − r)!
4.Combination without Repetition (Binomial coefficient)
nCr = n!/r!(n − r)!

```python
#Python (>2.6) 
import itertools as it
display(list(it.permutations([1, 2, 3])))
a = len(list(it.permutations([1, 2, 3])))
print(a)

#Python (<2.6) 
def all_perms(elements):
    if len(elements) <=1:
        #yield is a keyword that is used like return, except the function will return a 
        #generator[mygenerator = (x*x for x in range(3))].
        yield elements
    else:
        for perm in all_perms(elements[1:]):
            for i in range(len(elements)):
                # nb elements[0:1] works in both string and list contexts
                yield perm[:i] + elements[0:1] + perm[i:]
                
display(list(all_perms([1,2,3])))

#nPr function in python
import math
def nPr(n,r):
    f = math.factorial
    return f(n) // f(n-r)

#nCr function in python
import math
def nCr(n,r):
    f = math.factorial
    return f(n) // f(r) // f(n-r)

print(nCr(3,2))
```

## 3 Bysian Models 

Bayes Equation 

\begin{align}
 P( A | X ) = & \frac{ P(X | A) P(A) } {P(X) } \\\\[5pt]
& \propto P(X | A) P(A)\;\; (\propto \text{is proportional to } )
\end{align}


Example : 
 is my code bug free 

<ol> 
  <li>P(A) is prior probablity that code has no bugs  </li>
  <li>P(~A) is prior probablity that code has bugs =  1 - P(A) </li>
  <li>P(X) is the probablity of passing tests given that A (no bugs) or ~A (possible bugs)  </li>
  <li>P( A | X ) is posterior probability of no bugs, given our debugging tests  X  pass</li>
  <li>P( X | A ) is the probability of passing X tests, given that our code has no bugs </li>
  
</ol> 


\begin{align}
P(X ) & = P(X \text{ and } A) + P(X \text{ and } \sim A) \\\\[5pt]
 & = P(X|A)P(A) + P(X | \sim A)P(\sim A)\\\\[5pt]
& = P(X|A)p + P(X | \sim A)(1-p)
\end{align}

P(X|A), the probability that the code passes  X  tests given there are no bugs? Well, it is equal to 1.

P(X|~A), the probability that the code passes  X  tests given there are bugs? take a moderate guess = 0.5

assume that 
P(A) = p
P(~A) = 1 - p

\begin{align}
P(A | X) & = \frac{1\cdot p}{ 1\cdot p +0.5 (1-p) } \\\\
& = \frac{ 2 p}{1+p}
\end{align}

```python
%matplotlib inline
%pylab
import scipy.stats as stats
import numpy as np
import matplotlib as matplot 
from matplotlib import pyplot as plt
from scipy.stats import binom
plt.style.use('ggplot') # Make the graphs a bit prettier famous in R


p = np.linspace(0, 1, 50) ##prior belief
plt.plot(p, 2 * p / (1 + p), color="#348ABD", lw=3)
plt.xlabel("Prior, $P(A) = p$")
plt.ylabel("Posterior, $P(A|X)$, with $P(A) = p$")
```

```python
%matplotlib inline
#%pylab
import pymc as pm
import scipy.stats as stats
import numpy as np
import pandas as pd
import matplotlib as matplot 
from matplotlib import pyplot as plt
from scipy.stats import poisson
plt.style.use('ggplot') # Make the graphs a bit prettier famous in R
figsize(12.5, 10)

fdata = pd.read_csv('data/txtdata.csv')

m_fdata = mean(fdata.values)
c_fdata = fdata.count()[0]

alpha = 1.0 / m_fdata  # Recall count_data is the
                                  # variable that holds our txt counts
'''
#create new Instances of desired distributions
hint mu =  lambda #Poission coefficient 
'''   
mu_1 = pm.Exponential("mu_1", alpha)
mu_2 = pm.Exponential("mu_2", alpha)
tau = pm.DiscreteUniform("tau", lower=0, upper=c_fdata)
# each class generates random variables 
#print("Random output:", lambda_1.random(), lambda_2.random(), tau.random())


'''
@pm.deterministic is a decorator that tells PyMC this is a deterministic function.
'''
@pm.deterministic
def mu_(tau=tau, mu_1=mu_1, mu_2=mu_2):
    out = np.zeros(c_fdata)
    out[:tau] = mu_1  # lambda before tau is lambda1
    out[tau:] = mu_2  # lambda after (and including) tau is lambda2
    return out

#observed = True to tell PyMC that this should stay fixed in our analysis. 
observation = pm.Poisson("obs", mu_, value=fdata.values, observed=True)

model = pm.Model([observation, mu_1, mu_2, tau])

'''
MCMC  starts working on our model
'''
mcmc = pm.MCMC(model)
mcmc.sample(40000, 10000, 1)

'''
MCMC  generates samples for mu_1 mu_2 tau 
'''

mu_1_samples = mcmc.trace('mu_1')[:]
mu_2_samples = mcmc.trace('mu_2')[:]
tau_samples = mcmc.trace('tau')[:]

'''
plot histogram of the samples:
'''
#hist for mu_1
ax = plt.subplot(311)
ax.set_autoscaley_on(False)
plt.hist(mu_1_samples, histtype='stepfilled', bins=30, alpha=0.85,
         label="posterior of $\mu_1$", color="#A60628", normed=True)
plt.legend(loc="upper left")
plt.title(r"""Posterior distributions of the variables
    $\mu_1,\;\mu_1,\;\tau$""")
plt.xlim([15, 30])
plt.xlabel("$\mu_1$ value")

#hist for mu_2
ax = plt.subplot(312)
ax.set_autoscaley_on(False)
plt.hist(mu_2_samples, histtype='stepfilled', bins=30, alpha=0.85,
         label="posterior of $\mu_2$", color="#7A68A6", normed=True)
plt.legend(loc="upper left")
plt.xlim([15, 30])
plt.xlabel("$\mu_2$ value")

#hist for tau
ax = plt.subplot(313)
ax.set_autoscaley_on(False)
ax.hist(tau_samples,label="posterior of $\tau_2$",bins=c_fdata, color="g")
leg = plt.legend()
leg.get_frame().set_alpha(0.4)
plt.autoscale(tight=True)

```

```python
figsize(12.5, 5)
# tau_samples, lambda_1_samples, lambda_2_samples contain
# N samples from the corresponding posterior distribution
N = tau_samples.shape[0] #30000


expected_texts_per_day = np.zeros(c_fdata)
for day in range(0, c_fdata):
    # ix is a bool index of all tau samples corresponding to
    # the switchpoint occurring prior to value of 'day'
    ix = day < tau_samples
    #print(ix) ix is a binary array according to the condition
    # Each posterior sample corresponds to a value for tau.
    # for each day, that value of tau indicates whether we're "before"
    # (in the lambda1 "regime") or
    #  "after" (in the lambda2 "regime") the switchpoint.
    # by taking the posterior sample of lambda1/2 accordingly, we can average
    # over all samples to get an expected value for lambda on that day.
    # As explained, the "message count" random variable is Poisson distributed,
    # and therefore lambda (the poisson parameter) is the expected value of
    # "message count".
    expected_texts_per_day[day] = (mu_1_samples[ix].sum()
                                   + mu_2_samples[~ix].sum()) / N
    print('----------------')
    print('day =' + str(day))
    print('sum of mu_1_samples =' + str(mu_1_samples[ix].sum()))
    print('sum of mu_2_samples =' + str(mu_2_samples[~ix].sum()))
    print('N =' + str(N))
    print(expected_texts_per_day[day])
    print('----------------')

#plot the expectation line   
plt.plot(range(c_fdata), expected_texts_per_day, lw=4, color="#E24A33",
         label="expected number of text-messages received")

#plot the original oberved data
plt.bar(np.arange(c_fdata), fdata.values, color="#348ABD", alpha=0.65,
        label="observed texts per day")

plt.legend(loc="upper left");
```
